```python
number = int(input("Enter the number: "))

factorial = 1
for index in range(1,number+1):
    factorial *= index
    
print(factorial)
```

```python
number = int(input("Enter the number: "))
palindrome = number
reverse = 0
while number != 0:
    reverse = reverse*10 + number % 10
    number = number // 10
    
if palindrome == reverse:
    print("pallendrome")
else:
    print("Not a pallendrome")
```
