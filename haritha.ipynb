```python
salary= 500000
if salary <= 400000:
    print("0%")
elif 400000 < salary <= 800000:
    print("5%")
elif 800000 < salary <= 1200000:
    print("10%")
elif 1200000 < salary <= 1600000:
    print("15%")
elif 1600000 < salary <= 2000000:
    print("205")
elif 2000000 < salary <= 2400000:
    print("25%")
elif salary > 2400000:
    print("30%")
```
