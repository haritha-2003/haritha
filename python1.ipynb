```python
#datatypes
int_var = 10
print("integer = ",int_var)
print(type(int_var))
float_var = 10.3
print("float = ",float_var)
print(type(float_var))
comp_var = 5+4j
print("complex = ",comp_var)
print(type(comp_var))
my_str = "hari"
print("string = ",my_str)
print(type(my_str))
bool_var = True
print("boolean = ",bool_var)
print(type(bool_var))
my_set = {4,6,3,2,1}
print("set = ",my_set)
print(type(my_set))
dict_var = {'name':'haritha','age':22}
print("dictionary = ",dict_var)
print(type(dict_var))
tup_var = (4,6,7,2)
print("tuple = ",tup_var)
print(type(tup_var))
my_range = range(3)
print("range = ",my_range)
print(type(my_range))
my_list = [1,2,"hari",3.5,True]
print("list = ",my_list)
print(type(my_list))
```
