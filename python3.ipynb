```python
#operators
#modulo operator
int1 = 50
int2 = 10
print(int1 % int2)
float1 = 4.0
float2 = 2.5
print(float1 % float2)
```

```python
#floor division
int1 = 20
int2 = 6
print(int1 // int2)
float1 = 12.6
float2 = 6.0
print(float1 // float2)
```

```python
#exponent
int1 = 5
int2 = 3
print(int1 ** int2)
float1 = 3.5
float2 = 2.0
print(float1 ** float2)
```

```python
#float to int
float1 = 36.96
int1 = int(float1)
print(int1)
```

```python
#int to float
int1 = 45
float1 = float(int1)
print(float1)
```

```python
#string to int and float
str1 = '56'
int1 = int(str1)
print(int1)
str2 = '43.5'     
float1 = float(str2)
print(float1)
```
