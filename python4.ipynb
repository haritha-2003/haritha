```python
#enclosing scope
def outer():
    name = "haritha"
    def inner():
        print(name)
    inner()
outer()
```

```python
def outer1():
    sur = "haritha"
    print(name1)
    def inner1():
        name1 = "bandaru" 
        print(sur)    
    inner1()
outer1()
```

```python
def outer2():
    name1 = "haritha"     
    sur1 = " " 
    def inner2():
        nonlocal sur1
        sur1 = "bandaru"         
        print(name1)
    inner2()
    print(sur1)
outer2()
```

```python
#global scope
var3 = 25
def fun():
    print(var3)
fun()
print(var3)
```

```python
num = 10
def func():
    global num1
    num1 = 5
    print(num)
    print(num1)
func()
print(num)
```
