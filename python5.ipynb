```python
#comparison operator
print(5>4)
print(2<7)
print(4==3)
print(5!=4)
print(5>=6)
print(7<=6)
```

```python
#if condition
age = 20
if age <=18:
    print("not eligible")
else:
    print("eligible")
```

```python
age = int(input("what is the age:"))
if age <=18:
    print("not eligible")
else:
    print("eligible")
```

```python
marks = int(input("marks obtained:"))
if marks >= 90:
    print("grade A")
elif marks >= 80:
    print("grade B")
elif marks >= 70:
    print("grade C")
elif marks >= 60:
    print("grade D")
else:
    print("pass")
```

```python
#nested conditional statements
is_citizen = True 
age = 20
if is_citizen:
    if age > 18:
        print("eligible")
else:
    print("not eligible")
```

```python
#truthy and falsy
print(bool(False))
print(bool(0))
print(bool(0.0))
print(bool(""))
print(bool(True))
print(bool(1))
print(bool('ha'))
```

```python
#and operator
is_citizen = True
age = 22
if is_citizen and age>= 18:
    print("you are eligible")
else:
    print("you are not eligible")
```

```python
#or operator
is_citizen = True
age = 12
if is_citizen or age>= 18:
    print("you are eligible")
else:
    print("you are not eligible")
```

```python
#not operator
print(not '')
print(not 0)
print(not 'hello')
print(not True)
print(not False)
```
