```python
#len()
numbers = [1,2,3,4,5,6]
print(len(numbers))
names = ['haritha','shayam','lohith','nishanth']
print(len(names))
```

```python
#updating
marks = [80,85,90,75,78,60]
marks[1] = 56
print(marks)
```

```python
#remove
devops = ['terraform','docker','k8s','git','jenkins']
del devops[3]
print(devops)
```

```python
#in keyword
language = ['c','python','java','c++','html']
print('python' in language)
print('jbf' in language)
```

```python
#nested lists
string = [1,'fjns',['adc',2,'wefr']]
print(string)
print(string[2])
print(string[2][1])
print(string[2][0])
```

```python
#unpacking
details = ['haritha',22,'female']
name, age, gender = details
print(name)
print(age)
print(gender)
print(details)
```

```python
details = ['haritha',22,'female']
name, *rest = details
print(name)
print(rest)
print(*rest)
```

```python
#slicing
names = ['haritha','sruthi','anusha','prasanna','uma','anitha','sharmi']
print(names[1:6])
print(names[0:])
print(names[:7])
print(names[1:7:2])
print(names[0::2])
print(names[::-1])
```
