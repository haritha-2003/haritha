```python
#append
nums = [1,2,3,4,5]
nums.append(6)
print(nums)
print(nums.append(9))
```

```python
odds = [1,3,5,7,9]
evens = [2,4,6,8,10]
odds.append(evens)
print(odds)
```

```python
#insert
num = [1,2,3,4,5]
num.insert(3,6)
print(num)
print(num.insert(2,7))
```

```python
#remove
marks = [90,65,98,87,56,75]
marks.remove(87)
print(marks)
```

```python
#pop
nums = [1,3,6,7,8,4]
nums.pop(2)
print(nums)
```

```python
numer = [1,2,3,5,6]
numer.pop()
print(numer)
```

```python
#clear
nums = [1,2,3]
nums.clear()
print(nums)
```

```python
#sort
nums = [6,23,5,7,45,8]
nums.sort()
print(nums)
```

```python
#sorted
nums = [3,56,23,2,65,15]
nums2 = sorted(nums)
print(nums)
print(nums2)
```

```python
#reverse
numbers = [12,45,32,56,78,15]
numbers.reverse()
print(numbers)
strngs = ["hbf","hwb","iwe","qwe","oiq"]
strngs.reverse()
print(strngs)
```

```python
#index
strs = ['c','c++','java','python','html']
print(strs.index('python'))
print(strs.index())
```
